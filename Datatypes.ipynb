```python
#integer example

a = 10
b = 5
#printing integers
print(a)
print(b )
#arithmetic operations with integers
sum_result = a + b
print (sum_result)
```

```python
    #float example

x = 10.5
y = -3.14
#printing floats
print(x)
print(y)
#arithmetic operations with floats
div_result = x/2
print(div_result)
```

```python
#complex number example

c = 2 + 3j
#printimg complex number
print(c)
```

```python
# String example
name = "Prnathi"
greeting = 'Hello, World!'

# Printing strings
print(name)
print(greeting)
# String concatenation
full_greeting = greeting + " " + name
print(full_greeting)
```

```python

# List example
fruits = ["apple", "banana", "cherry"]

# Printing list
print(fruits)  # Output: ['apple', 'banana', 'cherry']

# Accessing list elements
print(fruits[1])  # Output: banana

# Adding an element to the list
fruits.append("date")
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date']
```

```python
# Tuple example

coordinates = (10, 20)

# Printing tuple
print(coordinates)

# Accessing tuple elements
print(coordinates[0])
```

```python
# Boolean example
is_active = True
is_logged_in = False

# Printing boolean values
print(is_active)
print(is_logged_in)

# Boolean operations
result = is_active and is_logged_in
print(result)
```

```python
#set example

unique_numbers = {1,2,3,4}

#printing set

print(unique_numbers)

#addind an item to the set

unique_numbers.add(2)
unique_numbers.add(5)

print (unique_numbers)
```

```python
#dictionary example

person = {"name":"john","age":30}

#print dictionary

print(person)

#accessing dictionary values

print(person["name"])

#modifying dictionary values

person["age"] = 31
print(person)

#adding new key value pairs

person["city"] = "new york"

print(person)
```
